## # Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/iris.csv')
df.shape
```

```python
df.iloc[::30]
```

```python
y = pd.get_dummies(df['class'])
y.shape
```

```python
x = df.drop(labels='class', axis=1)
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=42)

f'{xtrain.shape}, {ytrain.shape}, {xtest.shape}, {ytest.shape}'
```

## # Model

```python
import keras
from keras.models import Sequential
from keras.backend import clear_session
from keras.layers import Dense, Dropout
```

```python
def model():
    clear_session()
    
    model = Sequential()
    model.add(Dense(units=4, activation='relu', input_dim=xtrain.shape[1]))
    model.add(Dense(units=4, activation='relu'))
    model.add(Dense(units=3, activation='softmax'))
    print(model.summary())
    model.compile(optimizer='adam', loss='categorical_crossentropy',
                  metrics=['categorical_accuracy'])
    return model
```

```python
clf = model()
```

```python
clf.fit(x=xtrain, y=ytrain, validation_data=(xtest, ytest), epochs=2000)
```

## # Evaluation

```python
from sklearn.metrics import classification_report
```

```python
pred = clf.predict(xtest) > 0.5 # 0.5 is our treshold
```

```python
_ = classification_report(y_pred=pred, y_true=ytest)
print(_)
```
