## # Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/games.csv')
df.shape, df.columns
```

```python
def _value_counts(df):
    for c in df.columns:
        print(f'# {c}:\n{df[c].value_counts()}\n\n')


def _describe(df):
    for c in df.columns:
        print(f'# {c}:\n{df[c].describe()}\n\n')


def _isnull(df):
    for c in df.columns:
        cnt = df[pd.isnull(df[c])].shape[0]
        if cnt != 0:
            print(f'# {c}: {cnt} rows')    
```

```python
_describe(df)
```

```python
_isnull(df)
```

```python
_value_counts(df)
```

```python
df.drop(labels=['Other_Sales', 'Global_Sales', 'Developer'], axis=1, inplace=True)
df.shape
```

```python
df.dropna(axis=0, inplace=True)
df.shape
```

```python
df = df[df['NA_Sales'] > 1]
df = df[df['EU_Sales'] > 1]
df.shape
```

```python
game_names = df['Name']
```

```python
df.drop(labels='Name', axis=1, inplace=True)
df.shape
```

```python
df['User_Score'] = df['User_Score'].apply(float)
```

```python
y = df[['NA_Sales', 'EU_Sales', 'JP_Sales']]
y.shape
```

```python
y_na = df['NA_Sales']
y_eu = df['EU_Sales']
y_jp = df['JP_Sales']
```

```python
x_dummies = df.drop(labels=['NA_Sales', 'EU_Sales', 'JP_Sales'], axis=1).copy()
x_dummies.shape
```

```python
_cat_columns = ['Platform', 'Genre', 'Publisher', 'Rating']
```

```python
for col in _cat_columns:
    dummies = pd.get_dummies(df[col])
    print(f'# {col}. Adding {dummies.shape[1]} columns')
    x_dummies = pd.concat(objs=[x_dummies, dummies], axis=1)
```

```python
x_dummies.drop(labels=_cat_columns, axis=1, inplace=True)
x_dummies.shape, x_dummies.columns
```

```python
x_dummies.head(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x_dummies, y, test_size=0.25, random_state=42)

f'{xtrain.shape}, {ytrain.shape}, {xtest.shape}, {ytest.shape}'
```

## # Model

```python
from keras.backend import clear_session
from keras.models import Model, Sequential
from keras.layers import Dense, Dropout, Activation, Input
```

### ## output dim = 3

```python
def model_dim3():
    clear_session()
    
    model = Sequential()
    model.add(Dense(units=158, activation='relu', input_dim=x_dummies.shape[1]))
    model.add(Dense(units=158, activation='relu'))
    model.add(Dense(units=3, activation='linear'))
    model.compile(optimizer='adam', loss='mean_squared_error',
                  metrics=['mean_absolute_error'])
    print(model.summary())
    
    return model
```

```python
reg_dim3 = model_dim3()
```

```python
reg_dim3.fit(x=xtrain, y=ytrain, validation_data=(xtest, ytest), epochs=2000)
```

```python
pred = reg_dim3.predict(xtest)
```

```python
ytest.mean() - pred.mean()
```

```python
ytest.std() - pred.std()
```

### ## output dim = 1

```python
def model_dim1():
    clear_session()
    
    layer_input = Input(shape=(60,))
    
    layer_h1 = Dense(units=(60), activation='sigmoid')(layer_input)
    layer_h2 = Dense(units=(30), activation='sigmoid')(layer_h1)
    
    layer_o1 = Dense(units=(1), activation='linear', name='NA')(layer_h2)
    layer_o2 = Dense(units=(1), activation='linear', name='EU')(layer_h2)
    layer_o3 = Dense(units=(1), activation='linear', name='JP')(layer_h2)
    
    model = Model(inputs=layer_input, outputs=[layer_o1, layer_o2, layer_o3])
    model.compile(optimizer='adam', loss='mean_squared_error', metrics='mean_absolute_error')
    print(model.summary())
    
    return model
```

```python
reg_dim1 = model_dim1()
```

```python
reg_dim1.fit(x=x_dummies, y=[y_na, y_eu, y_jp], epochs=2000)
```

```python
pred_na, pred_eu, pred_jp = reg_dim1.predict(x_dummies)
```

```python
y_na.mean() - pred_na.mean()
```

```python
y_eu.mean() - pred_eu.mean()
```

```python
y_jp.mean() - pred_jp.mean()
```
